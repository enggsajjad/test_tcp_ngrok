```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import csv
```

```python


bad_words = ['SSH-2.0','GET','Host','Connection','User','Accept','Cookie']

#replace trash from the file
with open('tcpdata.log') as oldfile, open('newtcpdata.log', 'w') as newfile:
    for line in oldfile:
        if not any(bad_word in line for bad_word in bad_words):
            newfile.write(line);


```

```python
snr = pd.read_csv('newtcpdata.log')
accData=snr.values.tolist()
accData=np.asarray([np.asarray(elem) for elem in accData])
snr['p1'] = snr['pm25']
snr['p2'] = snr['pm10']
snr['t'] = snr['temp']
snr['h'] = snr['hum']
snr['a'] = snr['atm']
snr['m1'] = snr['m1']


PM25 = np.array(snr['p1'])
PM10 = np.array(snr['p2'])
Temp = np.array(snr['t'])
Hum = np.array(snr['h'])
Atm = np.array(snr['a'])
m = np.linspace(0,np.size(Temp),np.size(Temp))


figure,ax=plt.subplots(1)
#ax.plot(m, P1,'r')
ax.plot(Temp,'r')
ax.set_xlabel(r'measurements',fontsize=15)
ax.set_ylabel(r'Temperature',fontsize=15)

#plt.xlim(0, 300)
plt.ylim(1, 12)


figure,ax=plt.subplots(1)
ax.plot(Hum,'g')
ax.set_xlabel(r'measurements',fontsize=15)
ax.set_ylabel(r'Humidity',fontsize=15)

figure,ax=plt.subplots(1)
ax.plot(Atm,'b')
ax.set_xlabel(r'measurements',fontsize=15)
ax.set_ylabel(r'Atm. Pressure',fontsize=15)

figure,ax=plt.subplots(1)
ax.plot(PM25)
ax.set_xlabel(r'measurements',fontsize=15)
ax.set_ylabel(r'PM25 Particles',fontsize=15)

figure,ax=plt.subplots(1)
ax.plot(PM10)
ax.set_xlabel(r'measurements',fontsize=15)
ax.set_ylabel(r'PM10 Particles',fontsize=15)
```
