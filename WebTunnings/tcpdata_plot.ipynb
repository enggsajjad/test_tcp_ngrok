```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import csv
```

```python


bad_words = ['SSH-2.0','GET','Host','Connection','User','Accept','Cookie']

#replace trash from the file
with open('tcpdata.log') as oldfile, open('newtcpdata.log', 'w') as newfile:
    for line in oldfile:
        if not any(bad_word in line for bad_word in bad_words):
            newfile.write(line);


```

```python
snr = pd.read_csv('newtcpdata.log')
accData=snr.values.tolist()
accData=np.asarray([np.asarray(elem) for elem in accData])
snr['p1'] = snr['pm25']
snr['p2'] = snr['pm10']
snr['t'] = snr['temp']
snr['h'] = snr['hum']
snr['a'] = snr['atm']
snr['m4'] = snr['m4']


PM25 = np.array(snr['p1'])
PM10 = np.array(snr['p2'])
Temp = np.array(snr['t'])
Hum = np.array(snr['h'])
Atm = np.array(snr['a'])
m4 = np.array(snr['m4'])
m = np.linspace(0,np.size(Temp),np.size(Temp))

figure, ((ax1, ax2,ax3), (ax4, ax5,ax6)) = plt.subplots(2, 3,figsize=(3*7,2*6))
ax1.plot(Temp,'r')
ax1.set_xlabel(r'measurements',fontsize=15)
ax1.set_ylabel(r'Temperature',fontsize=15)
#plt.xlim(0, 300)
#plt.ylim(1, 12)


ax2.plot(Hum,'--g')
ax2.set_xlabel(r'measurements',fontsize=15)
ax2.set_ylabel(r'Humidity',fontsize=15)

ax3.plot(Atm,'b')
ax3.set_xlabel(r'measurements',fontsize=15)
ax3.set_ylabel(r'Atm. Pressure',fontsize=15)


ax4.plot(PM25,'c')
ax4.set_xlabel(r'measurements',fontsize=15)
ax4.set_ylabel(r'PM25 Particles',fontsize=15)


ax5.plot(PM10,'k')
ax5.set_xlabel(r'measurements',fontsize=15)
ax5.set_ylabel(r'PM10 Particles',fontsize=15)

ax6.plot(m4,'y')
ax6.set_xlabel(r'measurements',fontsize=15)
ax6.set_ylabel(r'm4',fontsize=15)
```
